```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
sns.set_style("darkgrid")

#Lectura de archivos y asignación de índices

antiguedad_saldos = pd.read_csv('Antiguedad_Saldos.csv',index_col = 'FACTURA')
devoluciones = pd.read_csv('Devoluciones.csv',index_col='CVE_DOC')
facturacion = pd.read_csv('Facturacion.csv',index_col='CVE_DOC')
gastos2021 = pd.read_csv('Gastos_Costos_2021.csv')
gastos2022 = pd.read_csv('Gastos_Costos_2022.csv')
gastos2023 = pd.read_csv('Gastos_Costos_2023.csv')
credito = pd.read_csv('Notas_Credito.csv',index_col='CVE_DOC')
precios_productos = pd.read_csv('Precios_Productos.csv',index_col='CVE_DOC')
```

```python
#Eliminación de columnas innecesarias

gastos2021 = gastos2021.drop(['Unnamed: 0'], axis=1)
gastos2022 = gastos2022.drop(['Unnamed: 0'], axis=1)
gastos2023 = gastos2023.drop(['Unnamed: 0'], axis=1)
```

## Dataframe precios_productos

**DIAGRAMA DE CAJA Y BIGOTE**

```python
precios_productos.plot.box(figsize=(5,5), rot=90)
```

**HISTOGRAMA**

```python
fig, ax = plt.subplots(2,2)
fig.subplots_adjust(hspace=0.4, wspace=0.3)
ax[0,0].hist(precios_productos.CANT)
ax[0,1].hist(precios_productos.COSTO_UNITARIO_CALCULADO)
ax[1,0].hist(precios_productos.COSTO_TOTAL_CALCULADO)
ax[1,1].hist(precios_productos.MARGEN_TOTAL_CALCULADO)
ax[0,0].set_title('CANT')
ax[0,1].set_title('COSTO_UNITARIO_CALCULADO')
ax[1,0].set_title('COSTO_TOTAL_CALCULADO')
ax[1,1].set_title('MARGEN_TOTAL_CALCULADO')
```

**MAPA DE CALOR**

```python
precios_productos=precios_productos.drop(['CVE_ART'],axis=1)
```

```python
import seaborn as sns

ax = sns.heatmap(precios_productos, annot = True, cmap = 'plasma_r')
plt.show()
```
